```python
import pandas as pd
import os

pd.set_option('display.max_columns', 500)
```

```python
data_path = os.getcwd()[:-len('Notebooks')]+'Data/'
os.listdir(data_path)
```

```python
data = pd.read_excel(data_path + 'DIRECTORIO PUEBLA 2 copy.xlsx')
data.head()
```

```python
df = data[data.columns[:13]]
df.head()
```

```python
mas50trabajadores = df[df['Rango de Empleados'] == '51 a 250'].copy()
de11a50trabajadores = df[df['Rango de Empleados'] == '11 a 50'].copy()
de0a10trabajadores = df[df['Rango de Empleados'] == '0 a 10'].copy()
mayora250 = df[df['Rango de Empleados'] == 'Mayor a 250'].copy()
otras = df[(df['Rango de Empleados'] !=  '51 a 250') & (df['Rango de Empleados'] !=  '11 a 50') & (df['Rango de Empleados'] !=  '0 a 10') & (df['Rango de Empleados'] !=  'Mayor a 250')].copy()
```
